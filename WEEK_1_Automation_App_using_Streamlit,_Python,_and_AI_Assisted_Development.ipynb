```python
# ============================================
# Cell 2: Import Required Libraries
# ============================================

# Import all required libraries
import pandas as pd
import numpy as np
from faker import Faker
import os

# Create Faker object for fake sample data
fake = Faker()
```

```python
# ============================================
# Cell 3: Create Sample HR CSV Files
# ============================================

# Define number of records
num_records = 100

# -------------------------------
# Create resumes.csv
# -------------------------------
resumes = pd.DataFrame({
    "candidate_name": [fake.name() for _ in range(num_records)],
    "skills": np.random.choice([
        "Python, SQL, Machine Learning",
        "Excel, Communication, HR Analytics",
        "Python, Power BI, SQL",
        "Recruitment, Communication, Excel",
        "Machine Learning, Python, Deep Learning"
    ], num_records),
    "experience_years": np.round(np.random.uniform(0, 8, num_records), 1),
    "education": np.random.choice(["MBA", "B.Tech", "MCA", "B.Sc", "M.Com"], num_records)
})

# Save file
resumes.to_csv("resumes.csv", index=False)

# -------------------------------
# Create employee_attrition.csv
# -------------------------------
attrition = pd.DataFrame({
    "employee_id": [f"EMP{1000+i}" for i in range(num_records)],
    "employee_name": [fake.name() for _ in range(num_records)],
    "department": np.random.choice(["HR", "IT", "Finance", "Marketing", "Operations"], num_records),
    "attendance_percentage": np.random.randint(50, 100, num_records),
    "performance_score": np.random.randint(1, 6, num_records),
    "overtime_hours": np.random.randint(0, 25, num_records),
    "years_at_company": np.random.randint(1, 11, num_records),
    "monthly_income": np.random.randint(25000, 120000, num_records),
    "attrition_risk": np.random.choice(["Low", "Medium", "High"], num_records)
})

# Save file
attrition.to_csv("employee_attrition.csv", index=False)

# -------------------------------
# Create employee_emails.csv
# -------------------------------
emails = pd.DataFrame({
    "employee_name": [fake.name() for _ in range(num_records)],
    "email": [fake.email() for _ in range(num_records)],
    "role": np.random.choice(["HR Executive", "Manager", "Analyst", "Developer", "Team Lead"], num_records)
})

# Save file
emails.to_csv("employee_emails.csv", index=False)

# -------------------------------
# Create performance_data.csv
# -------------------------------
performance = pd.DataFrame({
    "employee_id": [f"EMP{2000+i}" for i in range(num_records)],
    "employee_name": [fake.name() for _ in range(num_records)],
    "department": np.random.choice(["HR", "IT", "Finance", "Marketing", "Operations"], num_records),
    "score": np.random.randint(1, 6, num_records),
    "feedback": np.random.choice([
        "Excellent performer",
        "Good progress",
        "Needs improvement",
        "Highly committed",
        "Average performance"
    ], num_records)
})

# Save file
performance.to_csv("performance_data.csv", index=False)

# Print success message
print("✅ Sample CSV files created successfully!")
print(os.listdir())
```

```python
# ============================================
# Cell 4: Create resume_filter.py
# ============================================

# Create Python code for resume filtering utility
resume_filter_code = '''
import pandas as pd

def filter_resumes(df, required_skills):
    """
    Filter resumes based on required skills.
    df -> resumes dataframe
    required_skills -> list of skills to search
    """
    pattern = "|".join(required_skills)
    filtered_df = df[df["skills"].str.contains(pattern, case=False, na=False)]
    return filtered_df
'''

# Write the utility code into a Python file
with open("resume_filter.py", "w") as f:
    f.write(resume_filter_code)

print("✅ resume_filter.py created successfully!")
```

```python
# ============================================
# Cell 5: Create attrition_analysis.py
# ============================================

# Create Python code for attrition analysis utility
attrition_code = '''
import pandas as pd

def analyze_attrition(df):
    """
    Identify employees with high attrition risk
    using rule-based logic.
    """
    high_risk = df[
        (df["attendance_percentage"] < 70) &
        (df["performance_score"] <= 2)
    ]
    return high_risk
'''

# Write the utility code into a Python file
with open("attrition_analysis.py", "w") as f:
    f.write(attrition_code)

print("✅ attrition_analysis.py created successfully!")
```

```python
# ============================================
# Cell 6: Create dashboard_utils.py
# ============================================

# Create Python code for dashboard summary utility
dashboard_code = '''
import pandas as pd

def get_dashboard_summary(df):
    """
    Generate summary statistics for HR dashboard.
    """
    summary = {
        "Average Attendance": round(df["attendance_percentage"].mean(), 2),
        "Average Performance": round(df["performance_score"].mean(), 2),
        "Average Overtime": round(df["overtime_hours"].mean(), 2),
        "High Attrition Count": df[df["attrition_risk"] == "High"].shape[0]
    }
    return summary
'''

# Write the utility code into a Python file
with open("dashboard_utils.py", "w") as f:
    f.write(dashboard_code)

print("✅ dashboard_utils.py created successfully!")
```

```python
# ============================================
# Cell 7: Create email_generator.py
# ============================================

# Create Python code for email generation utility
email_code = '''
import pandas as pd

def generate_email(row):
    """
    Generate personalized email text for each employee.
    """
    email_text = f"""
Dear {row['employee_name']},

We are pleased to connect with you in your role as {row['role']}.
Thank you for your valuable contribution to the organization.

We look forward to your continued success.

Best regards,
HR Team
"""
    return email_text
'''

# Write the utility code into a Python file
with open("email_generator.py", "w") as f:
    f.write(email_code)

print("✅ email_generator.py created successfully!")
```

```python
# ============================================
# Cell 8: Create performance_report.py
# ============================================

# Create Python code for performance report utility
performance_code = '''
import pandas as pd

def classify_performance(score):
    """
    Classify employee performance based on score.
    """
    if score >= 4:
        return "High Performer"
    elif score >= 3:
        return "Average Performer"
    else:
        return "Needs Improvement"
'''

# Write the utility code into a Python file
with open("performance_report.py", "w") as f:
    f.write(performance_code)

print("✅ performance_report.py created successfully!")
```

```python
# ============================================
# Cell 9: Create app.py
# ============================================

# Create the main Streamlit app file
app_code = '''
import streamlit as st
import pandas as pd

from resume_filter import filter_resumes
from attrition_analysis import analyze_attrition
from dashboard_utils import get_dashboard_summary
from email_generator import generate_email
from performance_report import classify_performance

# --------------------------------------------
# Streamlit page settings
# --------------------------------------------
st.set_page_config(
    page_title="HR Automation App",
    layout="wide",
    initial_sidebar_state="expanded"
)

# --------------------------------------------
# App title and description
# --------------------------------------------
st.title("📊 HR Automation App")
st.write("This app demonstrates HR automation use cases using Streamlit and Python.")

# --------------------------------------------
# Main page menu instead of sidebar
# --------------------------------------------
menu = st.selectbox(
    "Choose a module",
    [
        "Home",
        "Resume Screening",
        "Attrition Analysis",
        "HR Dashboard",
        "Email Generator",
        "Performance Report"
    ]
)

st.markdown("---")

# --------------------------------------------
# Home page
# --------------------------------------------
if menu == "Home":
    st.header("Welcome")
    st.write("Select a module from the dropdown above to start using the HR automation app.")
    st.write("Modules included:")
    st.write("- Resume Screening")
    st.write("- Attrition Analysis")
    st.write("- HR Dashboard")
    st.write("- Email Generator")
    st.write("- Performance Report")

# --------------------------------------------
# Resume Screening page
# --------------------------------------------
elif menu == "Resume Screening":
    st.header("Resume Screening Tool")
    st.write("Upload a CSV file containing candidate resumes.")

    uploaded_file = st.file_uploader(
        "Upload resumes.csv",
        type=["csv"],
        key="resume_uploader"
    )

    if uploaded_file is not None:
        df = pd.read_csv(uploaded_file)
        st.subheader("Uploaded Resume Data")
        st.dataframe(df.head())

        skills_input = st.text_input(
            "Enter required skills separated by commas",
            "Python, SQL, Machine Learning"
        )
        required_skills = [skill.strip() for skill in skills_input.split(",")]

        if st.button("Filter Resumes"):
            filtered_df = filter_resumes(df, required_skills)
            st.subheader("Shortlisted Candidates")
            st.dataframe(filtered_df)

            if filtered_df.empty:
                st.warning("No candidates matched the given skills.")
            else:
                st.success(f"{len(filtered_df)} candidate(s) shortlisted.")

# --------------------------------------------
# Attrition Analysis page
# --------------------------------------------
elif menu == "Attrition Analysis":
    st.header("Employee Attrition Analysis")
    st.write("Upload employee attrition data to identify high-risk employees.")

    uploaded_file = st.file_uploader(
        "Upload employee_attrition.csv",
        type=["csv"],
        key="attrition_uploader"
    )

    if uploaded_file is not None:
        df = pd.read_csv(uploaded_file)
        st.subheader("Uploaded Attrition Data")
        st.dataframe(df.head())

        if st.button("Analyze Attrition Risk"):
            high_risk_df = analyze_attrition(df)
            st.subheader("High Attrition Risk Employees")
            st.dataframe(high_risk_df)

            if high_risk_df.empty:
                st.success("No high-risk employees found based on current rules.")

# --------------------------------------------
# HR Dashboard page
# --------------------------------------------
elif menu == "HR Dashboard":
    st.header("HR Dashboard Summary")
    st.write("Upload employee attrition data to generate dashboard metrics.")

    uploaded_file = st.file_uploader(
        "Upload employee_attrition.csv for dashboard",
        type=["csv"],
        key="dashboard_uploader"
    )

    if uploaded_file is not None:
        df = pd.read_csv(uploaded_file)
        st.subheader("Uploaded Dashboard Data")
        st.dataframe(df.head())

        summary = get_dashboard_summary(df)

        col1, col2, col3, col4 = st.columns(4)
        col1.metric("Average Attendance", summary["Average Attendance"])
        col2.metric("Average Performance", summary["Average Performance"])
        col3.metric("Average Overtime", summary["Average Overtime"])
        col4.metric("High Attrition Count", summary["High Attrition Count"])

        st.subheader("Attrition Risk Distribution")
        st.bar_chart(df["attrition_risk"].value_counts())

        if "department" in df.columns:
            st.subheader("Department-wise Employee Count")
            st.bar_chart(df["department"].value_counts())
        else:
            st.warning("Column 'department' not found in uploaded file.")

# --------------------------------------------
# Email Generator page
# --------------------------------------------
elif menu == "Email Generator":
    st.header("Automated Personalized Email Generator")
    st.write("Upload employee email data to generate personalized messages.")

    uploaded_file = st.file_uploader(
        "Upload employee_emails.csv",
        type=["csv"],
        key="email_uploader"
    )

    if uploaded_file is not None:
        df = pd.read_csv(uploaded_file)
        st.subheader("Uploaded Email Data")
        st.dataframe(df.head())

        if st.button("Generate Emails"):
            df["generated_email"] = df.apply(generate_email, axis=1)
            st.subheader("Generated Emails")
            st.dataframe(df[["employee_name", "email", "role", "generated_email"]])

# --------------------------------------------
# Performance Report page
# --------------------------------------------
elif menu == "Performance Report":
    st.header("Performance Report Generator")
    st.write("Upload performance data to generate performance classification.")

    uploaded_file = st.file_uploader(
        "Upload performance_data.csv",
        type=["csv"],
        key="performance_uploader"
    )

    if uploaded_file is not None:
        df = pd.read_csv(uploaded_file)
        st.subheader("Uploaded Performance Data")
        st.dataframe(df.head())

        if st.button("Generate Performance Report"):
            df["performance_level"] = df["score"].apply(classify_performance)
            st.subheader("Performance Report")
            st.dataframe(
                df[[
                    "employee_id",
                    "employee_name",
                    "department",
                    "score",
                    "feedback",
                    "performance_level"
                ]]
            )
'''

# Write the Streamlit code into app.py
with open("app.py", "w") as f:
    f.write(app_code)

print("✅ app.py created successfully!")
```
